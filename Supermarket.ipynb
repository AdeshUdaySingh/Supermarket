## Creating facts and dimension table

```python
customer_dim = df[['Customer type']].drop_duplicates().reset_index(drop=True)
customer_dim['customer_id'] = customer_dim.index + 1
customer_dim
```

```python
gender_dim = df[['Gender']].drop_duplicates().reset_index(drop=True)
gender_dim['gender_id'] = gender_dim.index + 1
gender_dim
```

```python
product_dim = df[['Product line', 'Unit price', 'Rating']].drop_duplicates().reset_index(drop=True)
product_dim['product_id'] = product_dim.index + 1
product_dim
```

```python
branch_dim = df[['City']].drop_duplicates().reset_index(drop=True)
branch_dim['branch_id'] = branch_dim.index + 1
branch_dim
```

```python
payment_dim = df[['Payment']].drop_duplicates().reset_index(drop=True)
payment_dim['payment_id'] = branch_dim.index + 1
payment_dim
```

```python
df['Date'] = pd.to_datetime(df['Date'])
date_dim = df[['Date']].drop_duplicates().reset_index(drop=True)
date_dim['year'] = date_dim['Date'].dt.year
date_dim['month'] = date_dim['Date'].dt.month
date_dim['day'] = date_dim['Date'].dt.day
date_dim['date_id'] = date_dim.index + 1
date_dim
```

```python
time_dim = df[['Time']].drop_duplicates().reset_index(drop=True)
time_dim['hour'] = pd.to_datetime(time_dim['Time']).dt.hour
time_dim['minute'] = pd.to_datetime(time_dim['Time']).dt.minute
time_dim['time_id'] = time_dim.index + 1
time_dim
```

```python
# Merge the original dataframe with the dimension tables to get foreign keys
fact_table = df.merge(customer_dim, on=['Customer type']) \
               .merge(gender_dim, on=['Gender']) \
               .merge(product_dim, on=['Product line', 'Unit price', 'Rating']) \
               .merge(branch_dim, on=['City']) \
               .merge(date_dim, on='Date') \
               .merge(time_dim, on='Time') \
               .merge(payment_dim, on='Payment')

# Select only the relevant columns for the fact table
fact_table = fact_table[['customer_id', 'gender_id', 'product_id', 'branch_id', 'date_id', 'time_id', 'payment_id',
                         'Quantity', 'Tax 5%', 'Total', 'cogs', 'gross income']]
fact_table
```

#### What is the effect of different pricing strategies (unit price) on total sales?

```python
df['Unit price'].describe()
```

```python
def label_ratings(row):
    col_value = row['Unit price']  # Access the value in the 'Unit price' column for the current row
    if col_value >= 10 and col_value <= 25:
        return 'Low'
    elif col_value >= 26 and col_value <= 50:
        return 'Mid'
    elif col_value >= 51 and col_value <= 75:
        return 'High'
    else:
        return 'Premium'

# Apply the function row-wise
df['Price Range'] = df.apply(label_ratings, axis=1)
df
```

```python
revenue_by_unit_price = df.groupby('Price Range').agg({'Total' : 'sum'}).sort_values(by = 'Total', ascending = False).reset_index()
revenue_by_unit_price
```

```python
revenue_by_unit_price_bar = plt.bar(revenue_by_unit_price['Price Range'], revenue_by_unit_price['Total'])

for bar in revenue_by_unit_price_bar :
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha = 'center', va = 'bottom')
    
plt.title('Revenue by Price Range')
plt.show()
```

## Sales Performance

#### What is the total sales (or revenue) generated each month?

```python
query = """
SELECT 
CASE
    WHEN d.month = 1 THEN 'Jan' || ' ' || d.year
    WHEN d.month = 2 THEN 'Feb' || ' ' || d.year
    WHEN d.month = 3 THEN 'Mar' || ' ' || d.year
END AS Month,
ROUND(SUM(s.Total), 2) AS Revenue
FROM sales s
JOIN date d ON s.date_id = d.date_id
GROUP BY Month;
"""

result = pd.read_sql(query, con=engine)
result
```

#### Which branch or city is performing the best in terms of sales?

```python
query = """
SELECT b.City, ROUND(SUM(s.Total), 2) AS Revenue
FROM sales s
JOIN branch b ON s.branch_id = b.branch_id
GROUP BY b.City
ORDER BY Revenue DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

#### Which product lines generate the most revenue?

```python
query = """
SELECT p.Product_line, ROUND(SUM(s.Total), 2) AS Revenue
FROM sales s
JOIN product p ON s.product_id = p.product_id
GROUP BY p.Product_line
ORDER BY Revenue DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

#### What payment method is most popular among customers?

```python
query = """
SELECT p.Payment, ROUND(SUM(s.Total), 2) AS Revenue
FROM sales s
JOIN payment p ON s.payment_id = p.payment_id
GROUP BY p.Payment
ORDER BY Revenue DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

# Customer Behavior

#### How much revenue is generated by "Member" vs "Normal" customers?

```python
query = """
SELECT c.Customer_type, ROUND(SUM(s.Total), 2) AS Revenue
FROM sales s
JOIN customer c ON s.customer_id = c.customer_id
GROUP BY c.Customer_type
ORDER BY Revenue DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

#### How do sales differ between male and female customers?

```python
query = """
SELECT c.Gender, ROUND(SUM(s.Total), 2) AS Revenue
FROM sales s
JOIN customer c ON s.customer_id = c.customer_id
GROUP BY c.Gender
ORDER BY Revenue DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

#### What is the average transaction size for different customer types?

```python
query = """
SELECT c.Customer_type, ROUND(AVG(s.Total), 2) AS Average_Spent
FROM sales s
JOIN customer c ON s.customer_id = c.customer_id
GROUP BY c.Customer_type
ORDER BY Average_Spent DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

# Product Performance

#### What are the top-selling products in terms of quantity and revenue?

```python
# I have already calculated the top product line based on revenue.
query = """
SELECT p.product_line, SUM(s.quantity) AS Quantity
FROM sales s
JOIN product p ON s.product_id = p.product_id
GROUP BY p.product_line
ORDER BY Quantity DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

#### Which product lines are most profitable based on gross income and gross margin?

```python
# I have already calculated the top product line based on revenue.
query = """
SELECT p.product_line, ROUND(SUM(s.gross_income), 2) AS Gross_Income
FROM sales s
JOIN product p ON s.product_id = p.product_id
GROUP BY p.product_line
ORDER BY Quantity DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

#### Which product lines are more popular in specific cities or branches?

```python
query = """
SELECT b.city, p.product_line, ROUND(SUM(s.Total), 2) AS Revenue
FROM sales s
JOIN branch b ON s.branch_id = b.branch_id
JOIN product p ON s.product_id = p.product_id
GROUP BY b.city, p.product_line
ORDER BY b.city, Revenue DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

# Customer Satisfaction

#### What is the average customer rating for each product line?

```python
query = """
SELECT Product_line, ROUND(AVG(Rating), 2) AS Average_Rating
FROM Product
GROUP BY Product_line
ORDER BY Average_Rating DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

#### Which branch has the highest average product ratings?

```python
query = """
SELECT b.city, ROUND(AVG(p.rating), 2) AS Average_Rating
FROM sales s
JOIN branch b ON s.branch_id = b.branch_id
JOIN product p ON s.product_id = p.product_id
GROUP BY b.city
ORDER BY Average_Rating DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

# Time-based Analysis

#### At what time of day does the supermarket generate the highest sales/revenue?

```python
query = """
SELECT t.hour, ROUND(SUM(s.Total), 2) AS Revenue
FROM sales s
JOIN time t ON s.time_id = t.time_id
GROUP BY t.hour
ORDER BY Revenue DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

#### Which days of the month have the highest sales volume?

```python
query = """
SELECT d.day AS Day, SUM(s.quantity) AS Sum_of_Quantity, SUM(s.Total) AS Revenue
FROM sales s
JOIN date d ON s.date_id = d.date_id
GROUP BY d.day
ORDER BY Sum_of_Quantity DESC;
"""

result = pd.read_sql(query, con=engine)
result
```

# Operational Efficiency

#### What is the average order value for each customer or per branch?

```python
query = """
SELECT c.customer_id, c.customer_type, c.gender,
ROUND(AVG(SUM(s.Total) / SUM(s.quantity)), 2) AS Average_Order_Value
FROM sales s
JOIN customer c ON s.customer_id = c.customer_id
GROUP BY c.customer_id;
"""

result = pd.read_sql(query, con=engine)
result
```

#### What are the total COGS per branch, and how does it compare to gross income?

```python
query = """
SELECT b.city, SUM(s.cogs) AS Total_COGS, SUM(s.gross_income) AS Total_Income
FROM sales s
JOIN branch b ON s.branch_id = b.branch_id
GROUP BY b.city
ORDER BY Total_COGS DESC, Total_Income DESC;
"""

result = pd.read_sql(query, con=engine)
result
```
